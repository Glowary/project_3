```python
import csv
import numpy as np
import pandas as pd

from geopy.geocoders import Nominatim
```

```python
# Read CSV file into a DataFrame
df = pd.read_csv('Metro_median_sale_price_uc_sfrcondo_sm_month.csv')
df
```

```python
# Drop rows with any missing value across all columns
df_cleaned = df.dropna(how='any')
df_cleaned
```

```python
# How to count column index number
def excel_column_to_number(column_label):
    result = 0
    for char in column_label:
        result = result * 26 + (ord(char.upper()) - ord('A')) + 1
    return result

# Find column number for the first month
first_date_col = 'F'
column_number = excel_column_to_number(first_date_col)
print(f"The column number for '{first_date_col}' is: {column_number}")

# Find column number for the last month
last_date_col = 'GI'
column_number = excel_column_to_number(last_date_col)
print(f"The column number for '{last_date_col}' is: {column_number}")
```

```python
# Define the range of columns for average calculation
start_col_index = 6
end_col_index = 191

# Calculate average for each row and add column to DataFrame
df_cleaned['Average'] = df.iloc[:, start_col_index:end_col_index].mean(axis=1)
df_cleaned
```

```python
# Initialize a geolocator
geolocator = Nominatim(user_agent="city_locator")

# Get latitude and longitude for a city
def get_lat_long(city, state):
    location = geolocator.geocode(f"{city}, {state}")
    if location:
        return location.latitude, location.longitude
    else:
        return None, None
    
# Add new columns for latitude and longitude to the DataFrame
df_cleaned.loc[:, 'Latitude'], df_cleaned.loc[:, 'Longitude'] = zip(*df_cleaned.apply(lambda row: get_lat_long(row['RegionName'], row['StateName']), axis=1))
df_cleaned
```

```python
# Save as new file
df_cleaned.to_csv('updatedsaleprice.csv', index=False)
```

```python
# Read the updated CSV file into a DataFrame
df_updated = pd.read_csv('updatedsaleprice.csv')

# Drop columns F to GI containing months
columns_to_drop = df_updated.columns[5:191]  # F to GI columns
df_slim = df_updated.drop(columns=columns_to_drop).copy()

# Save the slim version
df_slim.to_csv('updatedsaleprice_slim.csv', index=False)
```
